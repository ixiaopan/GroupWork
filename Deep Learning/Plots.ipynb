```python
import numpy as np
import matplotlib.pyplot as plt

names = ['Only-BG-B', 'Only-BG-T', "No-FG"]
errors = [[44.2, 47.1, 60],\
        [41.1, 44.9, 40.5]]
X = np.arange(3)
fig = plt.figure(figsize = (5,3))
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, errors[0], color = '#2e72ad', width = 0.30, label = "Same as train")
ax.bar(X + 0.30, errors[1], color = '#d39234', width = 0.30, label = "Original")

x = np.arange(len(names))+0.15  # the label locations
ax.set_xticks(x)
ax.set_xticklabels(names,  fontsize =15)
ax.set_yticklabels(np.linspace(0, 100, 6), rotation=0, fontsize=15)
plt.ylim(0, 100)
plt.ylabel("Error", fontsize = 15)
plt.legend(fontsize = 15)
plt.ylabel("Test Accuracy",fontsize = 15)
plt.xlabel("Training dataset",fontsize = 15)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
names = ['Mixed-Rand', 'Original']
errors = [[ 53.78, 61.43],\
         [53.7, 66.28],
         [54.37, 84.01],
         [51.83, 79.83],
         [49.48, 93.6]]
X = np.arange(2)
fig = plt.figure(figsize = (5,3))
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, errors[0], color = "#2e72ad", width = 0.15, label = "Mixed-Next")
ax.bar(X + 0.15, errors[1], color = '#d39234', width = 0.15, label = "Mixed-Rand")
ax.bar(X + 0.3, errors[2], color = '#459b76', width = 0.15, label = "Mixed-Same")
ax.bar(X + 0.45, errors[3], color = '#c76526', width = 0.15, label = "Only-FG")
ax.bar(X + 0.6, errors[4], color = '#c17db8', width = 0.15, label = "Original")



x = np.arange(len(names))+0.30  # the label locations
ax.set_xticks(x)
ax.set_xticklabels(names, fontsize = 20)
plt.ylim(0, 100)
ax.set_yticklabels(range(0, 110, 20), rotation=0, fontsize=20)
plt.legend(frameon=True,bbox_to_anchor=(1.03, 1), loc='upper left', fontsize = 20)
plt.ylabel("Test Accuracy", fontsize = 20)
plt.xlabel("Training dataset", fontsize = 20)
```
